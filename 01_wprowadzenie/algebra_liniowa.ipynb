# Norma wektora

```python
v1 = np.array([0,4])
```

```python
np.linalg.norm(v1)
```

# Norma euklidesowa

```python
v2 = np.array([10,-10,13])
np.linalg.norm(v2)
```

# Odległość dwóch punktów na płaszczyźnie

```python
p = np.array([3,0])
q= np.array([0,4])
np.linalg.norm(p-q)
```


# Odległość dwóch punktów w przestrzeni Rn


```python
p = np.array([3,0,1])
q= np.array([0,4,2])

np.linalg.norm(p-q)
```

# Mnożenie macierzy

```python
X = np.array([[2,3,4],[4,5,6]])
Y = np.array([[1,2],[4,7],[89,9]])

np.dot(X,Y)
```

# Wyznacznik macierzy
